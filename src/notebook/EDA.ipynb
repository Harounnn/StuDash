# Data Preprocessing

***
## 1. Indicators dataframe
***

***
### We will be interested in few indicators so our dataframe can be cleaner and  shrink
***

```python
ed_ind = list()
for i in indicators_df['Topic']:
    if 'Education' in i:
        ed_ind.append(i)
ed_ind_arr = np.array(ed_ind)
ed_ind = list(np.unique(ed_ind_arr))

len(ed_ind)
```

```python
fltr = (indicators_df['Topic'] == ed_ind[0]) | (indicators_df['Topic'] == ed_ind[1]) | (indicators_df['Topic'] == ed_ind[2]) | (indicators_df['Topic'] == ed_ind[3])
print(indicators_df.shape)
new_indicators_df = indicators_df[fltr]
print(new_indicators_df.shape)
```

***
### We will also get rid of the details (most of the columns)
***

```python
new_indicators_df = new_indicators_df[['Topic','Indicator Name']]
```

```python
new_indicators_df.head(5)
```

```python
new_indicators_df.isnull().sum()['Indicator Name']
```

```python
new_indicators_df.shape[0] - len(new_indicators_df['Indicator Name'].unique())
```

```python
countries_df = countries_df[['Country Code','Short Name','Currency Unit','Region']]

print(countries_df.isnull().sum())

countries_df.dropna(inplace=True)

countries_df.head(5)
```

## 3. Main dataframe
***

```python
main_df = main_df.loc[:,:'2020']
main_df.isnull().sum()[['Country Name','Country Code','Indicator Name']]
```

```python
#Indicators present in our dataframe
len(main_df['Indicator Name'].unique())
```

***
### We want only indicators related to our topics saved previously in a list called 'ed_ind' that was used to shrink the indicators dataframe , so we will do so to clean our main dataframe
***

```python
df = main_df.merge(new_indicators_df,left_on='Indicator Name', right_on='Indicator Name', how='left')
df.head(5)
```

```python
df = df.dropna(axis=0,subset=['Topic']).sort_values('Topic').drop(['Country Name','Indicator Code'],axis=1)
df.head(5)
```

```python
#chaging the position of the topic column
cols = list(df.columns)
topic_str_index = cols.index('Topic')
if topic_str_index != 1:
    topic_str = cols[topic_str_index]
    cols.remove(topic_str)
    cols.insert(1,topic_str)
    df = df[cols]
df.head(5)
```

***
### After reviewing the UNICEF manual ,provided in the README file, I understand better the indicators and I end up selecting these two indicators and their derivatives in our dataframe:
**1. GAR : Gross Attendance Rate** <br>   
$$ GAR = \frac{S}{P}$$<br>$$S : Students \; attending \; that \; level \; of \; education$$<br>$$P : Population \; aged \; the \; official \; age \; for \; that \; level \; of \; education$$
**2. OSR : Out-of-school Rate**  <br> <br>
$$ OSR = \frac{P - A}{P}$$<br>$$P : Population \; aged \; the \; official \; age \; for \; that \; level \; of \; education$$<br>$$A : Children \; and \; young \; people \; aged \; the \; official \; age \; and \; attending \; that \; level \; of \; education$$
***

```python
selected_indicators = ['DHS: Gross attendance rate. Post Secondary. Female','DHS: Gross attendance rate. Post Secondary. Male',
                       'DHS: Gross attendance rate. Post Secondary','DHS: Proportion of out-of-school. Primary. Female',
                       'DHS: Proportion of out-of-school. Primary. Male','DHS: Proportion of out-of-school. Primary.']
df_filter2 = ((df['Indicator Name'] == selected_indicators[0]) | (df['Indicator Name'] == selected_indicators[1])| 
(df['Indicator Name'] == selected_indicators[2]) | (df['Indicator Name'] == selected_indicators[3]) | 
(df['Indicator Name'] == selected_indicators[4]) | (df['Indicator Name'] == selected_indicators[5]))
new_df = df[df_filter2]

#Percentage of null values
(new_df.isnull().sum() / new_df.shape[0]) * 100
```

***
### Due to the amount of null values we will work only on data between 1990 and 2015
***

```python
start_year = 1990
last_year = 2015

tmp = pd.melt(new_df,id_vars=['Indicator Name','Country Code'],
              value_vars=[str(i) for i in range(start_year,last_year + 1)]
              ,var_name='Year', value_name='Value')
tmp = tmp.dropna()
tmp = tmp.merge(countries_df,left_on='Country Code', right_on='Country Code', how='left')
tmp = tmp[['Indicator Name','Short Name','Country Code','Currency Unit','Region','Year','Value']]
tmp.head()
```

***
### Now we have this 'tmp' dataframe from which we will explore and visualize our data
***

***
# Exploration and Visualizations
***

***
### Which regions have the highest GAR ?
***

```python
q1 = tmp.groupby(['Indicator Name','Region'])['Value'].mean()['DHS: Gross attendance rate. Post Secondary'].sort_values(ascending=False)

q1 = q1.to_frame().reset_index()
q1['Value'] = round(q1['Value'],1)

fig = px.pie(q1, values='Value', names='Region', hole=0.05, hover_data={'Region':False},
             title=f'Mean Post Secandary GAR from {start_year} to {last_year}')
fig.show()
###
from IPython.display import Image, display
display(Image(filename='pieplot.png'))
```
