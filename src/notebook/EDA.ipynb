```python
l = list()
for i in df3['Topic']:
    if 'Education' in i:
        l.append(i)
arr = np.array(l)
l = list(np.unique(arr))
len(l)
```

```python
fltr = (df3['Topic'] == l[0]) | (df3['Topic'] == l[1]) | (df3['Topic'] == l[2]) | (df3['Topic'] == l[3])
print(df3.shape)
new_df3 = df3[fltr]
print(new_df3.shape)
```
