```python
#import the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#options
%matplotlib inline
pd.set_option('display.max_columns',None)
```

```python
main_df = pd.read_csv('Stats.csv')
countries_df = pd.read_csv('StatsCountry.csv')
indicators_df = pd.read_csv('StatsSeries.csv')

main_df.head()
```

# Data Preprocessing
***

***
## 1. Indicators dataframe
***

***
## We will be interested in few indicators so our dataframe will be cleaner and will shrink
***

```python
ed_ind = list()
for i in indicators_df['Topic']:
    if 'Education' in i:
        ed_ind.append(i)
ed_ind_arr = np.array(ed_ind)
ed_ind = list(np.unique(ed_ind_arr))

len(ed_ind)
```

```python
fltr = (indicators_df['Topic'] == ed_ind[0]) | (indicators_df['Topic'] == ed_ind[1]) | (indicators_df['Topic'] == ed_ind[2]) | (indicators_df['Topic'] == ed_ind[3])
print(indicators_df.shape)
new_indicators_df = indicators_df[fltr]
print(new_indicators_df.shape)
```

***
## We will also get rid of the details (most of the columns)
***

```python
new_indicators_df = new_indicators_df[['Topic','Indicator Name']]
```

```python
new_indicators_df
```

```python
new_indicators_df.isnull().sum()['Indicator Name']
```

```python
new_indicators_df.shape[0] - len(new_indicators_df['Indicator Name'].unique())
```

***
## The indicators dataframe is clean and good to go 
***

***
## 2. Countries dataframe
***

```python
countries_df = countries_df[['Country Code','Short Name','Currency Unit','Region']]

print(countries_df.isnull().sum())

countries_df.dropna(inplace=True)

countries_df
```

```python
print(f"Number of rows : {countries_df.shape[0]}")
print('------------------------------------------------------------------')

for col in countries_df.columns:
    print(f"Number of unique values in the {col} column is {len(countries_df[col].unique())}")
```

***
## Now the countries dataframe seems reasonable and good to go
***

***
## 3. Main dataframe
***

```python
main_df = main_df.loc[:,:'2020']

tmp_new_indicators_df = new_indicators_df.set_index('Indicator Name')

new_main_df = pd.merge(main_df,tmp_new_indicators_df,left_on='Indicator Name',right_index=True)

new_main_df
```

```python
#Number of indicators in our dataset = number of rows
new_main_df.shape[0]
```
